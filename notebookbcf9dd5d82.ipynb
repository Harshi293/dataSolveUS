```python
data1 = pd.read_csv("/kaggle/input/datasolve-us/train.csv")
test = pd.read_csv("/kaggle/input/datasolve-us/test.csv")
```

```python
test_doc=list(test['document_text'])
```

```python
data1['document_text']= data1['document_text'].str.lower()
```

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","â€“"]
for char in spec_chars:
    data1['document_text'] = data1['document_text'].str.replace(char, ' ')
```

```python
data1['document_text'] = data1['document_text'].str.replace(r'\b\w\b', '').str.replace(r'\s+', ' ')
data1['document_text'] = data1['document_text'].str.replace('\d+', '')
```

```python
data2 = data1.groupby('document_text')['cat_name'].apply(list)
data3=data2.to_frame().reset_index()
data4=tuple(data3['document_text'])
#data2= data1.groupby('document_text')
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np


mlb = MultiLabelBinarizer()
Y = mlb.fit_transform(labels)

# check size of indicator matrix
# print some rows 
Y[0]
Y.shape
# check classes
mlb.classes_

# check # of samples in each class
np.sum(Y, axis=0)
```

```python
import numpy as np
import xgboost
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC
from xgboost import XGBClassifier
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Multi-label classification- one vs. rest classifier
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.utils import class_weight

mlp=MLPClassifier(random_state=0, early_stopping=True, verbose=2)
sgd = SGDClassifier(random_state=0,loss='log',alpha=0.01,penalty='elasticnet')
lr = LogisticRegression(C=1.0)
svc = svm.LinearSVC()
nb = MultinomialNB()
xgb = XGBClassifier(scale_pos_weight=4)
X_train, X_test, Y_train, Y_test = train_test_split(\
                data4, Y, test_size=0.3, random_state=0)

classes_weights = class_weight.compute_sample_weight(
    class_weight='balanced',
    y=Y_train
)

classifier = Pipeline([
    ('tfidf', TfidfVectorizer(min_df=1,stop_words="english")),
    ('clf', OneVsRestClassifier(estimator=xgb))])

#kwargs = {classifier.steps[-1][0] + '__sample_weight': classes_weights}
#classifier.fit(X_train, Y_train, **kwargs)

classifier.fit(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report

predicted = classifier.predict(X_test)

predicted.shape
print("predicted:")
predicted[0:2]
print("actual:")
Y_test[0:2]

print(classification_report\
      (Y_test, predicted, target_names=mlb.classes_))
```

```python
#PREDICTING ON TEST DATA

# Multi-label classification- Performance report

from sklearn.metrics import classification_report

predicted = classifier.predict(test_doc)

predicted.shape
print("predicted:")
predicted[0:2]

#print(classification_report\
 #     (Y_test, predicted, target_names=mlb.classes_))
```

```python
#submission output
final_id=[]
label_list=[]
predicted_list=[]
ids=test['id']
for i in ids:
    for j in range(0,50):
        final_id.append(i)
    for k in mlb.classes_:
        label_list.append(k)
for l in predicted:
     for m in l: 
        predicted_list.append(m)
```

```python
zipped = list(zip(final_id,label_list,predicted_list))

output_df=pd.DataFrame(zipped, columns=['ID', 'LABEL', 'predictions'])
output_df["id"] = output_df["ID"].astype(str) + '_' + output_df["LABEL"]
```

```python
final_output = output_df[['id', 'predictions']]
final_output
```
